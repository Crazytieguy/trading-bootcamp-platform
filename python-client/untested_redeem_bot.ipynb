```python
ETFS = {
    'ghi_tw': {
        'golf_tw': 3,
        'hotel_tw': 2,
        'india_tw': 1
    },
    'def_tw': {
        'delta_tw': 1,
        'echo_tw': 1,
        'foxtrot_tw': 4
    },
    'abc_tw': {
        'alpha_tw': 2,
        'bravo_tw': 2,
        'charlie_tw': 2
    }
}

ALL_MARKETS = list(set(ETFS.keys()).union(*[set(etf_constituents.keys()) for etf_constituents in ETFS.values()]))

from dotenv import load_dotenv
from metagame import TradingClient
from metagame.websocket_api import Side
from typing import Literal
from metagame.websocket_api import CancelOrder, ClientMessage, CreateOrder, Side
from os import getenv


JWT = getenv("JWT")
API_URL = getenv("API_URL") 
# ACT_AS = int(getenv("ACT_AS", None))
ACT_AS=106 # arrbot



import time
while True:
    time.sleep(1)
    # client.state().portfolio.market_exposures
    with TradingClient(API_URL, JWT, ACT_AS) as client:
        market_name_to_id = client.state().market_name_to_id
        market_id_to_name = {v: k for k, v in market_name_to_id.items()}
        min_weighted_position = None
        for etf_data in ETFS:
            
            etf_id = market_name_to_id[etf_data]
            etf_constituents = ETFS[etf_data]
            exposures = client.state().portfolio.market_exposures
            for stock_name, stock_weight in etf_constituents.items():
                stock_id = market_name_to_id[stock_name]
                stock_exposures = [exposure for exposure in exposures if exposure.market_id == stock_id]
                if len(stock_exposures) > 1:
                    print(stock_exposures)
                    assert False, 'bizarre condition, num exposures > 1'
                if len(stock_exposures) < 1:
                    stock_position = 0
                elif len(stock_exposures) == 1:
                    stock_position = stock_exposures[0].position
                else:
                    assert False, 'bizarre condition'

                if min_weighted_position is None:
                    min_weighted_position = stock_position / stock_weight
                else:
                    min_weighted_position = min(min_weighted_position, stock_position * stock_weight)
            
            etf_exposures = [exposure for exposure in exposures if exposure.market_id == etf_id]
            if len(etf_exposures) > 1:
                print(etf_exposures)
                assert False, 'bizarre condition, etf exposures > 1'
            if len(etf_exposures) < 1:
                etf_position = 0
            elif len(etf_exposures) == 1:
                etf_position = etf_exposures[0].position
            else:
                assert False, 'bizarre condition'
                
            print(f'{min_weighted_position=}')
            print(f'{etf_position=}')

            # Check if signs are different between ETF position and minimum weighted position
            if min_weighted_position is not None and etf_position != 0:
                if (min_weighted_position > 0 and etf_position < 0) or (min_weighted_position < 0 and etf_position > 0):
                    print(f"Found opposing positions in {etf_data}:")
                    print(f"ETF position: {etf_position}")
                    print(f"Min weighted constituent position: {min_weighted_position}")
                    min_redeemable = min(abs(min_weighted_position), abs(etf_position))
                    # Redeem the minimum amount possible to reduce opposing positions
                    if etf_position > 0:
                        # If ETF position is positive but constituents negative, redeem ETF
                        print(f"Redeeming {min_redeemable} of {etf_data}")
                        client.redeem(etf_id, min_redeemable)
                    else:
                        # If ETF position is negative but constituents positive, redeem -amount
                        print(f"Redeeming {-min_redeemable} of {etf_data}")
                        client.redeem(etf_id, -min_redeemable)
```
